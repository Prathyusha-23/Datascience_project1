```python
import pandas as pd#series generated from a list
a=[1,7,2]
myvar = pd.Series(a)
print(a)
print(myvar)
print('------------------')
myvar2= pd.Series(a,index = ["x","y","z"])
print(myvar2)
print(myvar2["z"])
print('------------------')
calories={"day1":420,"day2":380,"day3":390}
myvar=pd.Series(calories)
print(myvar)
print(myvar["day2"])
print('------------------')
#series generated from scalar value
series=pd.Series(5,index=[0,1,2,3])
print(series)
print('------------------')
```

```python
import pandas as pd
#integer series
int_series=pd.Series([1,2,3,4],index=["a","b","c","d"])
print(int_series)
print('-----------------')
#float series
float_series=pd.Series([1.1,2.2,3.3,4.4])
print(float_series)
print('-----------------')
#String series
string_series=pd.Series(['a','b','c','d'])
print(string_series)
print('-----------------')
#label based indexing
data={'a':10,'b':20,'c':30}
series=pd.Series(data)
print(series)
print(series['a'])
print(series[['a','c']])
#Boolean indexing
data=[10,20,30,40,50]
series=pd.Series(data)
#access elements greater than 25
print(series[series>25])
print('---------------')
```

```python
data={'Name':['Alice','bob','Ave'],
      'Age':[25,30,35],
      'city':['New york','Los Angeles','Chicago']}
df=pd.DataFrame(data)
print(df)
print('------------------')
#selecting columns')
print(df.Name)
print(df['Name'])
print('-------------------')
print(df[['Name','city']])
print('----------------')
#select a single row by label
print("1.\n",df.loc[0])
print("2.\n",df.iloc[0])
print('--------------')
#conditional selection
print('conditional')
print(df[df['Age']>25])
print(df.loc[(df.Name=='bob')&(df.city=='LosAngeles')])
```

```python
import pandas as pd
data={
    'name':['Alice','Bob','charlie','Alice','Bob'],
    'age':[10,20,30,10,20],
    'city':['Newyork','francisco','los angeles','Newyork','francisco'],
    'salary':[70000,80000,120000,70000,80000]
}
df=pd.DataFrame(data)
grouped=df.groupby('name')
print(grouped.groups)
print("")
#calculate the sum of salaries for each group
salary_sum= grouped['salary'].sum()
print(salary_sum)
print("*")
age_mean=grouped['age'].mean()
print(age_mean)
print("------------------")
#Grouping by multiple columns
#group by 'name' and 'city'
grouped_multiple=df.groupby(['name','city'])
#calculate the sum of salaries for each group
salary_sum_multiple=grouped_multiple['salary'].sum()
print(salary_sum_multiple)
```

```python
#pandas merging,joining,concatenation of the dataframes
import pandas as pd
data1={'Name':['jai','princi','gaurav','anuj'],
       'Age':[27,24,22,32],
       'Address':['nagapur','kanpur','allahabad','karnataka'],
       'Qualification':['msc','ma','btech','phd']}
data2={'Name':['Abhi','sai','cool','jani'],
       'Age':[21,22,43,56],
       'Address':['hyd','vzg','nellore','ongole'],
       'Qualification':['B.Tech','B.A','B.Com','B.hons']}
df=pd.DataFrame(data1,index=[0,1,2,3])
df1=pd.DataFrame(data2,index=[4,5,6,7])
print(df,"\n\n",df1)
print("------------------")
#concatenation
frames=[df,df1]
result=pd.concat(frames)
print(result)
```

```python
import pandas as pd
data1={'Name':['jai','princi','gaurav','anuj'],
       'Age':[27,24,22,32],
       'Address':['nagapur','kanpur','allahabad','karnataka'],
       'Qualification':['msc','ma','btech','phd']}
data2={'Name':['Abhi','sai','cool','jani'],
       'Age':[21,22,43,56],
       'Address':['hyd','vzg','nellore','ongole'],
       'Qualification':['B.Tech','B.A','B.Com','B.hons']}
df=pd.DataFrame(data1,index=[0,1,2,3])
df1=pd.DataFrame(data2,index=[2,3,6,7])
print(df,"\n\n",df1)
res1=pd.concat([df,df1])
print(res1)
res2=pd.concat([df,df1],axis=1,join='inner')
print(res2)
```

```python
#joining of 2 datframes by ignoring the indexes
import pandas as pd
data1={'Name':['jai','princi','gaurav','anuj'],
       'Age':[27,24,22,32],
       'Address':['nagapur','kanpur','allahabad','karnataka'],
       'Qualification':['msc','ma','btech','phd']}
data2={'Name':['Abhi','sai','cool','jani'],
       'Age':[21,22,43,56],
       'Address':['hyd','vzg','nellore','ongole'],
       'Qualification':['B.Tech','B.A','B.Com','B.hons']}
df=pd.DataFrame(data1,index=[0,1,2,3])
df1=pd.DataFrame(data2,index=[2,3,6,7])
print(df,"\n\n",df1)
res=pd.concat([df,df1],ignore_index=True)
print(res)
```

```python
#concatenate a mix of series and dataframes
import pandas as pd
data1={'Name':['jai','princi','gaurav','anuj'],
       'Age':[27,24,22,32],
       'Address':['nagapur','kanpur','allahabad','karnataka'],
       'Qualification':['msc','ma','btech','phd']}
#convert the dictionary into datframe
df=pd.DataFrame(data1,index=[0,1,2,3])
#creating the series
s1=pd.Series([1000,2000,3000,4000],name='Salary')
#concatenate the series and dataframe
result=pd.concat([df,s1],axis=1)
print(result)
```

```python
#Merging the dataframes
import pandas as pd
data1={'key':['k0','k1','k2','k3'],
       'Name':['Jai','Princi','Gaurav','Anuj'],
       'Age':[27,24,22,32]}
data2={'key':['k0','k1','k2','k3'],
       'Address':['Nagpur','Kanpur','Allahabad','Kannauj'],
       'Qualification':['Btech','B.A','Bcom','B.hons']}
df=pd.DataFrame(data1)
df1=pd.DataFrame(data2)
res=pd.merge(df,df1,on='key')
print(res)
```

```python
import pandas as pd
data1={'key':['k0','k1','k2','k3'],
       'key1':['k0','k1','k0','k1'],
       'Name':['Jai','Princi','Gaurav','Anuj'],
       'Age':[27,24,22,32]}
data2={'key':['k0','k1','k2','k3'],
       'key1':['k0','k0','k0','k0'],
       'Address':['Nagpur','Kanpur','Allahabad','Kannauj'],
       'Qualification':['Btech','B.A','Bcom','B.hons']}
df=pd.DataFrame(data1)
df1=pd.DataFrame(data2)
print(df,"\n\n",df1)
res=pd.merge(df,df1,on=['key','key1'])
print(res)
```

```python
#mering:inner,outer,left,right
#By default inner join will come
df1={'name':['prathyusha','praveen','hari'],
     'email':['soda@gmail.com','chintu@gmail.com','hari@gmail.com'],
     'age':['18','17','19'],
     'city':['ongole','goa','punjab'],
     'address1':['gandhinagar','mm road','kurnool road']}
df2={'name':['soda','chintu','priya'],
     'email':['s@gmail.com','c@gmail.com','h@gmail.com'],
     'age':['18','17','19'],
     'city':['ongole','manipur','punjab'],
     'address2':['sujatha nagar','rajeev nagar','bustand']}
df=pd.DataFrame(df1)
df3=pd.DataFrame(df2)
print(df,"\n\n",df3)
res=pd.merge(df,df3,on='city',how='outer',indicator=True,suffixes=('_left','_right'))
print(res)
```
